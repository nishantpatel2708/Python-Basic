### Pattern 1

```python
n = 5

for i in range(n, 0, -1):
  for j in range(0, i):
    print(n, end=' ')

  print()
```

### Pattern 2

```python
n = 5

for i in range(n, 0, -1):
  for j in range(0, i + 1):
    print(j, end='  ')

  print()
```

### Pattern 6

```python
n = 7
p = 1

for i in range(1, n+1):
    
    for j in range(0, i):
        if j==0 or i==0:
            p = 1
        else:
            p = p * (i - j)//j
        print(p, end = " ")
    print()
```

### Pattern 7

```python
n = 5
for i in range(1, n + 1):  
    for j in range(1, n + 1):  
        if j <= i:  
            print(i, end=' ')  
        else:  
            print(j, end=' ')  
    print()  
```

### Pattern 8

```python
n = 9
for i in range(1, n):  
    for j in range(1, i + 1):  
        print(i * j, end='  ')  
    print()  
```

### Pattern 9

```python
n = 5  
k = 2 * n - 2  
for i in range(n, -1, -1):  
    for j in range(k, 0, -1):  
        print(end=" ")  
    k = k + 1  
    for j in range(0, i + 1):  
        print("*", end=" ")  
    print("")  
```

### Pattern 10

```python
n = 5
m = (2 * n) - 2  
for i in range(0, n):  
    for j in range(0, m):  
        print(end=" ")  
    m = m - 1    
    for j in range(0, i + 1):  
        print("* ", end=' ')  
    print(" ")  
```

### Pattern 12

```python
n = 5
for i in range(0, n):
    for j in range(0, i + 1):
        print("*", end=' ')
    print("\r")

for i in range(n, 0, -1):
    for j in range(0, i - 1):
        print("*", end=' ')
    print("\r")
```

### Pattern 14

```python
n = 5
for i in range(n, 0, -1):
    for j in range(n-i):
        print(" ", end="")
    for j in range(1, 2*i):
        print("*", end="")
    print()

for i in range(2, n+1):
    for j in range(n-i):
        print(" ", end="")
    for j in range(1, 2*i):
        print("*", end="")
    print()
```

### Pattern 15

```python
n = 14
print("*" * n, end="\n")
i = (n // 2) - 1
j = 2
while i != 0:
    while j <= (n - 2):
        print("*" * i, end="")
        print(" " * j, end="")
        print("*" * i, end="\n")
        i = i - 1
        j = j + 2
```
